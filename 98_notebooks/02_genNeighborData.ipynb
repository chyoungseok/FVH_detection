```python
import os, glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm

path_src = "/zdisk/users/ext_user_03/01_yschoi/project_01_FVH_detection/01_data/04_flair_preproc_slices_v02"
path_dst = "/zdisk/users/ext_user_03/01_yschoi/project_01_FVH_detection/01_data/04_flair_preproc_slices_v02_neighbor"

list_subjects = [f for f in os.listdir(path_src) if not(f.startswith('.')) and os.path.isdir(os.path.join(path_src, f))]
print(f"Number of subjects: {len(list_subjects)}")

for _subject in tqdm(list_subjects[:]):
    _path_img = glob.glob(os.path.join(path_src, _subject, "*padded.npy"))[0]
    _path_label = glob.glob(os.path.join(path_src, _subject, "*sliceLevel.npy"))[0]
    
    img = np.load(_path_img)
    label = np.load(_path_label)
    
    list_slice = []
    list_label = []
    if label.any():
        # FHV positive 존재
        idx_range = np.argwhere(label).ravel() # fhv가 존재하는 slice의 index (0-based)
    else:
        # FHB positive 부재
        idx_range = range(12, 16)
    for i in idx_range:
        _slice_3d = img[i-1:(i+1)+1]
        _label = label[i]
        
        list_slice.append(_slice_3d)
        list_label.append(_label)
    
    arr_slice = np.array(list_slice)
    arr_label = np.array(list_label)
    
    _path_save_subject = os.path.join(path_dst, _subject)
    os.makedirs(_path_save_subject, exist_ok=True)
    np.save(os.path.join(_path_save_subject, f"{_subject}_AxialSlices_padded.npy"), arr_slice)
    np.save(os.path.join(_path_save_subject, f"{_subject}_label_sliceLevel.npy"), arr_label)
    
    if label.any():
        print(f"=== {_subject} ===")
        print(arr_slice.shape, arr_label.shape)
        print(_path_save_subject)
        print()
    
```
